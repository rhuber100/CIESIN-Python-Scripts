```python
import os
from openpyxl import load_workbook
import re

# Directory containing the files
directory = '/content/drive/MyDrive/Summer work/Sources for GDP Datasets'

# Iterate through files in the directory
for filename in os.listdir(directory):
    # Check for files that have "Usa" as a standalone word (case insensitive)
    if re.search(r'\bUsa\b', filename, re.IGNORECASE):  # \b asserts a word boundary
        file_path = os.path.join(directory, filename)

        # Check if the file exists and is an Excel file
        if os.path.exists(file_path) and filename.endswith(('.xlsx', '.xls')):
            try:
                # Load the Excel file
                wb = load_workbook(file_path)

                # Get the names of all sheets
                sheet_names = wb.sheetnames

                # Sheets to keep (ensure correct casing)
                sheets_to_keep = ['Sheet1', 'Aggregated Data']

                # Delete sheets that are not in sheets_to_keep
                for sheet in sheet_names:
                    if sheet not in sheets_to_keep:
                        del wb[sheet]

                # Save the modified workbook
                wb.save(file_path)
                print(f'Modified: {filename}')

            except Exception as e:
                print(f'Error processing {filename}: {e}')  # Print the error message
        else:
            print(f'File not found or not an Excel file: {file_path}')
```

```python
import os
from openpyxl import load_workbook
import re

# Directory containing the files
directory = '/content/drive/MyDrive/Summer work/Sources for GDP Datasets'

# Iterate through files in the directory
for filename in os.listdir(directory):
    # Check for files that have "Gbr" as a standalone word (case insensitive)
    if re.search(r'\bGbr\b', filename, re.IGNORECASE):  # \b asserts a word boundary
        file_path = os.path.join(directory, filename)

        # Check if the file exists and is an Excel file
        if os.path.exists(file_path) and filename.endswith(('.xlsx', '.xls')):
            try:
                # Load the Excel file
                wb = load_workbook(file_path)

                # Get the names of all sheets
                sheet_names = wb.sheetnames

                # Delete the sheet named "Aggregated GDP Final" if it exists
                if "Aggregated GDP Final" in sheet_names:
                    del wb["Aggregated GDP Final"]
                    print(f'Deleted sheet "Aggregated GDP Final" from {filename}')

                # Save the modified workbook
                wb.save(file_path)
                print(f'Modified: {filename}')

            except Exception as e:
                print(f'Error processing {filename}: {e}')  # Print the error message
        else:
            print(f'File not found or not an Excel file: {file_path}')
```
