```python
import os
import pandas as pd

# Directory containing your files
input_dir = '/content/drive/MyDrive/Summer work/Sources for GDP Datasets'

# List all files in the directory
file_list = os.listdir(input_dir)

# Create a DataFrame to store the extracted region names
data = []
for file in file_list:
    if file.endswith('.xlsx') or file.endswith('.gsheet'):  # Include .gsheet files
        # Extract the region name from the file name
        # Example: "RegionName.xlsx" -> "RegionName"
        region_name = os.path.splitext(file)[0]  # Remove file extension
        data.append([file, region_name])

# Create DataFrame
df = pd.DataFrame(data, columns=['file_name', 'Region Name'])

# Save as CSV to use as the input table for the matching script
input_table_path = '/content/drive/MyDrive/Summer work/matching/input_table.csv'
df.to_csv(input_table_path, index=False)
print(f"Input table saved at {input_table_path}")
```

```python
import geopandas as gpd
import pandas as pd

# Path to your GeoJSON file
geojson_path = '/content/drive/MyDrive/Summer work/matching/geoBoundariesCGAZ_ADM2.geojson'

# Load the GeoJSON file using GeoPandas
geo_df = gpd.read_file(geojson_path)

# Combine 'shapeName' and 'shapeGroup' with 'shapeName' first and 'shapeGroup' second
geo_df['Region Name'] = geo_df['shapeName'] + ' ' + geo_df['shapeGroup']

# Extract relevant columns
matching_table = geo_df[['shapeGroup', 'Region Name']]

# Rename columns for clarity
matching_table.columns = ['ISO Code', 'Region Name']

# Save the matching table as CSV
matching_table_path = '/content/drive/MyDrive/Summer work/matching/matching_table.csv'
matching_table.to_csv(matching_table_path, index=False)

print(f"Matching table saved at {matching_table_path}")
```
