```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
from google.colab import userdata
```

```python
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python

import pandas as pd

# Create a list of region names

region_names = ["King's County NY", "Scarborough, Maine"]



output_folder = "/content/drive/MyDrive/Gemini Countries"

# Iterate over the region names and generate content for each
for region in region_names:
    response = model.generate_content(f"Create a two column table of the GDP breakdown of {region} by sectoral contributions from agriculture, industry, and services, no descriptions, percentages.")
    formatted_response = to_markdown(response.text)
    print(formatted_response.data)

    # Create a DataFrame with the formatted response
    # splitResponse = formatted_response.data.split('|')

    # columns = [splitResponse[0], splitResponse[1]]
    # data = filter(lambda x: x != '\n> ', splitResponse)
    # evenMoreFilteredData = filter(lambda x: x != '---', data)
    # thirdFilteredData = filter(lambda x: x != '> ', evenMoreFilteredData)
    # print(list(thirdFilteredData))
    df = pd.DataFrame({"GDP Breakdown": formatted_response.data}, index=[0])  # Use formatted_response.data  # Add a single-element index

    # Create the Excel file path with region name
    file_path = f"{output_folder}/{region.replace(' ', '_')}.xlsx"  # Replace spaces with underscores for a valid filename

    # Save the DataFrame to Excel
    df.to_excel(file_path, index=False)  # Don't save the index

```
