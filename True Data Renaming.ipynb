```python
import os
import pandas as pd

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Define file paths
folder_path = '/content/drive/MyDrive/Summer work/Sources for GDP Datasets'
spreadsheet_path = '/content/drive/MyDrive/Summer work/Master Table True Data.xlsx'  # Adjust for correct extension

# Load the master table spreadsheet without headers
df = pd.read_excel(spreadsheet_path, header=None)

# Iterate over the rows starting from row 2 (index 1 in pandas)
for index, row in df.iloc[1:].iterrows():  # Skip the first row (header)
    old_file_name = row[0]  # Column A (file name)
    new_file_name = row[1]  # Column B (shape name)

    # Check if new_file_name is NaN or not a string
    if pd.isna(new_file_name):
        print(f"Missing shape name for {old_file_name}, skipping...")
        continue

    # Convert new_file_name to string
    new_file_name = str(new_file_name)

    # Check if the old file has an extension, if not, default to '.xlsx'
    file_extension = old_file_name.split('.')[-1] if '.' in old_file_name else 'xlsx'

    # Construct old and new full file paths
    old_file_path = os.path.join(folder_path, old_file_name)
    new_file_path = os.path.join(folder_path, new_file_name + f'.{file_extension}')

    # Check if the old file exists
    if os.path.exists(old_file_path):
        try:
            # Rename the file, keeping the same extension
            os.rename(old_file_path, new_file_path)
            print(f"Renamed: {old_file_name} -> {new_file_name}.{file_extension}")

            # Update the file name in the DataFrame (Column A)
            df.at[index, 0] = new_file_name + f'.{file_extension}'
        except OSError as e:
            print(f"Error renaming {old_file_name}: {e}")
    else:
        print(f"File not found: {old_file_name}")

# Save the updated spreadsheet, keeping the same structure (no header)
df.to_excel(spreadsheet_path, index=False, header=False)

print("File renaming and spreadsheet update complete.")
```

```python
import pandas as pd

# Load the master table spreadsheet without headers
spreadsheet_path = '/content/drive/MyDrive/Summer work/Master Table True Data.xlsx'
df = pd.read_excel(spreadsheet_path, header=None)

# Iterate over the rows starting from row 2 (index 1 in pandas)
for index, row in df.iloc[1:].iterrows():  # Skip the first row (header)
    # Get the current file name and the shape name
    current_file_name = row[0]  # Column A (current file name)
    shape_name = row[1]  # Column B (shape name)

    # Check if the shape name is not NaN
    if pd.notna(shape_name):
        # Determine the file extension from the current file name
        file_extension = current_file_name.split('.')[-1] if '.' in current_file_name else 'gsheet'

        # Construct the new file name
        new_file_name = f"{shape_name}.{file_extension}"

        # Update the file name in the DataFrame (Column A)
        df.at[index, 0] = new_file_name

# Save the updated spreadsheet, keeping the same structure (no header)
df.to_excel(spreadsheet_path, index=False, header=False)

print("File names in the master table updated successfully.")
```
