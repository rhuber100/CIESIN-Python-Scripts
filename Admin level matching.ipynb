```python
# Import necessary libraries
import pandas as pd
import os
import re
from openpyxl import load_workbook
from google.colab import drive
from zipfile import BadZipFile

# Mount Google Drive
drive.mount('/content/drive')

# Paths for the Excel file and the source folder
input_excel_path = '/content/drive/MyDrive/Summer work/Country Admin Levels Table.xlsx'
source_folder_path = '/content/drive/MyDrive/Summer work/Sources for GDP Datasets'

# Load the Excel file as a DataFrame
admin_table = pd.read_excel(input_excel_path)  # Load the Excel file
admin_table.columns = ['ISO Code', 'Admin Level']  # Ensure the columns are named correctly

# Print available ISO codes for debugging
print("Available ISO Codes in Admin Table:")
print(admin_table['ISO Code'].unique())  # Display unique ISO codes

# Iterate through the Excel files in the source folder
for filename in os.listdir(source_folder_path):
    if filename.endswith('.xlsx'):
        # Check if "CAN" is a standalone ISO code in the filename
        if re.search(r'\bCAN\b', filename):
            # Extract the ISO code as 'CAN' based on the filename
            iso_code_in_file = 'CAN'
            print(f"Processing file: {filename}, Extracted ISO Code: {iso_code_in_file}")  # Debug print

            # Check if the extracted ISO code matches any in the admin table
            admin_level_row = admin_table[admin_table['ISO Code'].str.upper() == iso_code_in_file]

            if not admin_level_row.empty:  # If there's a match
                admin_level = admin_level_row.iloc[0]['Admin Level']  # Get the admin level

                # Open the Excel file and add the "Admin Level" information
                file_path = os.path.join(source_folder_path, filename)

                try:
                    workbook = load_workbook(file_path)
                    sheet = workbook.active

                    # Write the header and the administrative level value
                    sheet.cell(row=1, column=19, value="Admin Level")
                    sheet.cell(row=2, column=19, value=admin_level)

                    # Save the changes
                    workbook.save(file_path)

                    # Print confirmation message
                    print(f"Successfully modified: {filename}")

                except BadZipFile:
                    print(f"Skipping bad file: {filename} (not a valid Excel file)")
                except Exception as e:
                    print(f"An error occurred while processing {filename}: {e}")  # Catch any other exceptions
            else:
                print(f"No matching admin level found for ISO Code: {iso_code_in_file}")  # Debug print

print("Script completed successfully.")
```
